### Exploring the db_loader

```python
import os
import pandas as pd
import psycopg
from pgvector.psycopg import register_vector, Vector  # Vector is key!

DSN = os.getenv("AIPE_PG_DSN", "postgresql://aipe:aipe123@localhost:5433/aipe")

def query_df(sql, params=None):
    # Open a short-lived connection for each query
    with psycopg.connect(DSN) as con:
        register_vector(con)  # teach psycopg about pgvector
        return pd.read_sql(sql, con, params=params)
```

```python
cands = query_df("""
  SELECT candidate, COUNT(*) AS n
  FROM transcript_chunks
  GROUP BY 1
  ORDER BY 2 DESC
""")
cands
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("intfloat/multilingual-e5-large")

def embed_query(text: str) -> Vector:
    # E5 expects 'query: ' prefix and normalized output
    v = model.encode([f"query: {text}"], normalize_embeddings=True)[0].astype("float32").tolist()
    return Vector(v)  # wrap so Postgres sees a real 'vector', not float[]
  
```

```python
# 🔎 Your question & the specific candidate (use an exact value from Cell 2)
question  = "Dame una opinión controversial que haya tenido y que se refiera sobre su propio pensamiento y no de otra persona"
candidate = "Keiko Fujimori"     # <-- replace with a value that exists in your data
k = 5

qv = embed_query(question)

sql = """
SELECT id, candidate, text, start, "end", video_id,
       1 - (embedding <=> %s) AS score
FROM transcript_chunks
WHERE candidate = %s
ORDER BY embedding <=> %s
LIMIT %s
"""
df = query_df(sql, params=[qv, candidate, qv, k])
df[["candidate", "score", "text", "start", "end", "video_id"]]

df.to_csv('prueba.csv',index=False)
```
